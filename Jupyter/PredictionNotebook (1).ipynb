### Splitting the data for training and testing model

```python
df.shape
```

```python
df.head(2)
```

```python
col= (['itching','skin rash','continuous sneezing','joint pain','stomach pain','acidity',
             'ulcers on tongue','vomiting','burning micturition','spotting urination','fatigue','weight gain','anxiety',
              'restlessness','cough','high fever','breathlessness','dehydration','indigestion',
             'dark urine','nausea','back pain','constipation','yellowing of eyes','chest pain'])
```

```python
len(col)
```

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

#### If gives error re-run this particular cell

```python
from sklearn.decomposition import PCA

pca= PCA(n_components=25)
pca.fit(x)
```

```python
pca_x= pca.transform(x)
```

```python
pca_x= pd.DataFrame(pca_x, columns= col)
pca_x
```

```python
x_train, x_test, y_train, y_test=train_test_split(pca_x, y, random_state=42, test_size=0.3)
```

```python
print(f"X_TrainData: {x_train.shape}")
print(f"X_TestData: {x_test.shape}")
print(f"Y_TrainData: {y_train.shape}")
print(f"Y_TestData: {y_test.shape}")
```

            #We are done with cleaning the data and splitting it into training and testing 

# Model Building 

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

## Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb= GaussianNB()
```

```python
nb.fit(x_train,y_train)
pred1= nb.predict(x_test)
print(accuracy_score(pred1,y_test))
```

```python
print(classification_report(pred1,y_test))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
pred=rf.predict(x_test)
```

```python
print(accuracy_score(pred,y_test))
```

```python
print(classification_report(pred,y_test))
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
pred=dt.predict(x_test)
```

```python
print(accuracy_score(pred,y_test))
print('=========')
print(classification_report(pred,y_test))
```

## Support Vector Machine

```python
from sklearn.svm import SVC
svm= SVC(kernel='linear')
```

```python
svm.fit(x_train,y_train)
pred=svm.predict(x_test)
```

```python
print(accuracy_score(pred,y_test))
print('=========')
print(classification_report(pred,y_test))
```
